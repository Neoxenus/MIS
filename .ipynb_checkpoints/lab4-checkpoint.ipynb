## Варіант 5
$$\begin{equation}
  \left\{
  \begin{aligned}
     & 8.1x_1 + 8.5x_2 &= 72 \\
     & -3.2x_1 + 15.2x_2 &= 42 \\
     & 10.2x_1 + 8.8x_2 &= 205
  \end{aligned}
  \right.
\end{equation}$$

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.linalg import solve
```

```python
x1_values = np.linspace(0, 80, 400)
x12_values = (72 - 8.1 * x1_values) / 8.5
x22_values = (42 + 3.2 * x1_values) / 15.2
x32_values = (205 - 10.2 * x1_values) / 8.8
```

```python


x2_min = min(np.min(x12_values), np.min(x22_values), np.min(x32_values))
x2_max = max(np.max(x12_values), np.max(x22_values), np.max(x32_values))

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(x1_values, x12_values, color="red", linestyle="dashed", label='8.1x₁ + 8.5x₂ = 72')
plt.title('Графік лінії 8.1x₁ + 8.5x₂ = 72')
plt.legend()
plt.grid(True)
plt.ylim(x2_min, x2_max)


plt.subplot(1, 3, 2)


plt.plot(x1_values, x22_values,  color="blue", linestyle="dotted", label='-3.2x₁ + 15.2x₂ = 42')

plt.title('Графік лінії -3.2x₁ + 15.2x₂ = 42')
plt.legend()
plt.grid(True)
plt.ylim(x2_min, x2_max)

plt.subplot(1, 3, 3)
plt.plot(x1_values, x32_values, color="black", linestyle="dashdot", label='10.2x₁ + 8.8x₂ = 205')

plt.title('Графік лінії 10.2x₁ + 8.8x₂ = 205')
plt.legend()
plt.grid(True)
plt.ylim(x2_min, x2_max)

plt.tight_layout()
plt.show()
```

```python

# Побудуємо графік
plt.figure(figsize=(8, 16), dpi=100)

# Побудуємо лінії та зафарбуємо область перетину
plt.plot(x1_values, x12_values, label='f[0]', color='red', linestyle='dashed')
plt.plot(x1_values, x22_values, label='f[1]', color='blue', linestyle='dotted')
plt.plot(x1_values, x32_values, label='f[2]', color='black', linestyle='dashdot')


plt.fill_between(x1_values, x12_values, x2_max, color='gray', alpha=0.3)
plt.fill_between(x1_values, x32_values, x2_max, color='white', alpha=1.0)
plt.fill_between(x1_values, x22_values, x2_max, color='white', alpha=1.0)
# Додамо сітку з вказаним коліром та типом лінії
plt.grid(color='black', linestyle='--')

# Додамо підписи рівнянь прямих вздовж ліній
plt.annotate('8.1x₁ + 8.5x₂ = 72', (x1_values[0]-3, x12_values[0] - 3), rotation=-60, fontsize=8)
plt.annotate('-3.2x₁ + 15.2x₂ = 42', (x1_values[200], x22_values[200] + 1.5), rotation=15, fontsize=8)
plt.annotate('10.2x₁ + 8.8x₂ = 205', (x1_values[0], x32_values[0] - 5), rotation=-60, fontsize=8)

# Додамо підписи осей та назву графіку
plt.xlabel('x₁')
plt.ylabel('x₂')
plt.title('Перетин прямих')
plt.legend()

intersection_point_1 = solve([[8.1, 8.5], [-3.2, 15.2]], [72, 42])
intersection_point_2 = solve([[8.1, 8.5], [10.2, 8.8]], [72, 205])
intersection_point_3 = solve([[-3.2, 15.2], [10.2, 8.8]], [42, 205])
print(intersection_point_1)
print(intersection_point_2)
print(intersection_point_3)
plt.scatter(*intersection_point_1, color='black', marker='o', label=f'Intersection 1: {intersection_point_1}')
plt.scatter(*intersection_point_2, color='black', marker='o', label=f'Intersection 2: {intersection_point_2}')
plt.scatter(*intersection_point_3, color='black', marker='o', label=f'Intersection 3: {intersection_point_3}')
plt.annotate('Перетин f[1] та f[2]', (intersection_point_3[0] + 3, intersection_point_3[1] - 0.5), rotation=0, fontsize=10)
plt.annotate('Перетин f[0] та f[2]', (intersection_point_2[0] - 21, intersection_point_2[1] - 1), rotation=0, fontsize=10)
# Показати графік
plt.show()
```
